#### This is a basic demo of time series decomposition into components Level, Trend, Seasonality and Residuals

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.axes as ax
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# demo1: sleep dataset. Has level, no trend and fluctuates about the level
plt.rcParams['figure.figsize'] = [15, 4]
ts = pd.read_csv("sleepDuration.csv")
ts.plot()
plt.show()
```

```python
# decompose and plot components
plt.rcParams['figure.figsize'] = [15, 5]
ts_components = seasonal_decompose(ts, model='additive', period=4)
ts_components.plot()
plt.show()
```

```python
# demo2: air passengers dataset, has clear trend and multiplicative seasonality
ts = pd.read_csv('airline-passengers.csv', header=0, index_col=0)
ts.plot()
plt.show()
```

```python
plt.rcParams['figure.figsize']  = [12, 8]
ts_components = seasonal_decompose(ts, model='multiplicative',period=12)
ts_components.plot()
plt.show()
```
