**Create a bar plot showing the average value in millions of dollars per year**

```python
# CODE HERE

umtmvs.resample('Y').mean().head()
```

```python
umtmvs.resample('Y').mean().plot.bar(figsize=(12,6))
 # resample will sample the datatime index on the basis of given parameter
```

**What year had the biggest increase in mean value from the previous year's mean value? (Lots of ways to get this answer!)**

[HINT for a useful method](https://pandas.pydata.org/pandas-docs/version/0.21/generated/pandas.DataFrame.idxmax.html)

```python
# CODE HERE
```

```python
umtmvs_yrly = umtmvs.resample('Y').mean()
umtmvs_yrly.head()
```

```python
umtmvs_yrly_1 = umtmvs_yrly.shift(1)
umtmvs_yrly_1.head()
```

```python
umtmvs_change = umtmvs_yrly_1 - umtmvs_yrly
umtmvs_change.head()
```

```python
umtmvs_change['UMTMVS'].idxmax()
```

```python
umtmvs_change['UMTMVS']['2009-12-31']
```

```python
print(umtmvs_change['UMTMVS'].idxmax())
```

**Plot out the yearly rolling mean on top of the original data. Recall that this is monthly data and there are 12 months in a year!**

```python
# CODE HERE

umtmvs['Yearly_mean'] = umtmvs['UMTMVS'].rolling(window=12).mean()
umtmvs.tail()
```

```python
umtmvs.plot(figsize=(15,8))
```
